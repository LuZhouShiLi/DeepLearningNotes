```python
with segyio.open(waveform_path+fname, "r", strict=False) as f:
    data_win = []
    sta_name = []
     # 文件信息
    sample_rate = 1000000/f.header[0][117]
    # sample_rate_resampled = sample_rate * resample_rate
     # datetime_start = datetime_fromtraceheader(f.header[0])
    datetime_start = datetime.strptime(
    fname.split('.')[0], "%Y%m%d_%H%M%S")
    
    tnum = f.tracecount
    stanum = int(tnum/3)
    for i in range(int(tnum/3)):
        sta_name.append(get_name_from_int(f.header[i*3][61]))
    
    print(sta_name)
    print(stanum)


def get_name_from_int(num):
    nameint = num & 0xFFFFFFFF
    namebytes = nameint.to_bytes(4, byteorder='little', signed=False)
    return "".join(["%02X" % (i) for i in namebytes])[2:]
```
