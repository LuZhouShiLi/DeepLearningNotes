```python
import segyio
import numpy as np
import os

# SEGY
folder_path = '../SegyData'
# NPZ
save_path = '../NPZDATA'

# SEGY
for filename in os.listdir(folder_path):
    if filename.endswith('.sgy'):
        segy_file_path = os.path.join(folder_path, filename)
        
        # segyioSEGY
        with segyio.open(segy_file_path, 'r', ignore_geometry=True) as segy:
            # 
            num_traces = segy.tracecount  # 
            num_samples = segy.samples.size  # 
            
            # 
            # data = segyio.tools.cube(segy)
            data = segy.trace.raw[:]

            # data 
            data = data.reshape((num_traces,num_samples))

        
        # NPZ
        # npz
        npz_file_path = os.path.join(save_path, os.path.splitext(filename)[0] + '.npz')
        # npz
        np.savez(npz_file_path, data=data, num_traces=num_traces, num_samples=num_samples)
        
        print(f"Saved {filename} as {npz_file_path}")
```
